# STEP 2 -  Create X and y

```python
X = c.drop(['resale_price'], axis = 1)
y = c['resale_price']
```

# STEP - 3 Train test Split

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

train = []
test = []
cvs = []

for i in range(0,100):
    X_train,X_test,y_train,y_test = train_test_split(X,y,train_size = 0.8, random_state = i )
    lr.fit(X_train,y_train)
    ypred_train = lr.predict(X_train)
    ypred_test = lr.predict(X_test)
    train.append(lr.score(X_train,y_train))
    test.append(lr.score(X_test,y_test))
    cvs.append(cross_val_score(lr,X,y,cv = 5).mean())
    
em = pd.DataFrame({"Train":train,"Test":test,"CVS":cvs})
gm = em[(abs(em["Train"]-em["Test"])<=0.05) & (abs(em["Test"]-em["CVS"])<=0.05)]
print("Best Random State Number:",gm[gm["Test"]==gm["Test"].max()].index.to_list()[0])
```

**Linear Regression**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size = 0.8, random_state = 89)

lr.fit(X_train,y_train)
print("Coefficient = ", lr.coef_)
print("Intercept = ", lr.intercept_)
```

```python
from sklearn.metrics import r2_score

ypred_train = lr.predict(X_train)
ypred_test = lr.predict(X_test)

print("R2 for test data = ",r2_score(y_test,ypred_test))
print("R2 for train data = ",r2_score(y_train,ypred_train))
print("Cross Validation Score = ",cross_val_score(lr,X_train,y_train, cv = 5).mean())
```

# Applying ANN

```python
import tensorflow as tf
import keras
from keras.layers import Dense
from keras.models import Sequential
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
def ann_model_regressor():
    ann = Sequential()
    ann.add(Dense(input_dim = 10, units= 50, kernel_initializer = 'uniform',activation = 'relu' ))
    ann.add(Dense(units = 50 , kernel_initializer = 'uniform',activation = 'relu' ))
    ann.add(Dense(units = 50 , kernel_initializer = 'uniform',activation = 'relu' ))
    ann.add(Dense(units = 50 , kernel_initializer = 'uniform',activation = 'relu' ))
    ann.add(Dense(units = 50 , kernel_initializer = 'uniform',activation = 'relu' ))
    
    ann.add(Dense(units = 1 , kernel_initializer = 'uniform', activation = 'relu'))
    ann.compile(optimizer = 'adam', loss = 'mean_squared_error')
    return ann
```

```python
ann = ann_model_regressor()
```

```python
ann.fit(X_train, y_train, batch_size = 32, epochs = 100)
```

```python
ypred_tn =ann.predict(X_train)
ypred_ts =ann.predict(X_test)
```

```python
ts = pd.DataFrame(ypred_ts,columns = ['resale_price'])
ts
```

```python
from sklearn.metrics import mean_squared_error
print("MSE for train data = ",mean_squared_error(y_train,ypred_tn))
print("MSE for test data = ", mean_squared_error(y_test,ypred_ts))
```

```python
from sklearn.metrics import r2_score
print("R2 for train data = ",r2_score(y_train,ypred_tn))
print("R2 for test data = ",r2_score(y_test,ypred_ts))
```

```python
def biuld_cross_classifier():
    classifier = Sequential()
    classifier.add(Dense(input_dim = 10,units = 10, kernel_initializer = 'uniform', activation = 'relu'))
    classifier.add(Dense(units = 10, kernel_initializer = 'uniform', activation = 'relu'))
    classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'relu'))
    classifier.compile(optimizer = 'adam', loss = 'mean_squared_error',metrics = ['accuracy'] )
    return classifier
```

```python
from scikeras.wrappers import KerasRegressor
classifier = KerasRegressor(biuld_cross_classifier,batch_size = 32, epochs = 100)
```

```python
Cross_Validation = cross_val_score(classifier,X_train,y_train,cv = 5)
```
